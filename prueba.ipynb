```python
## Prueba de Notebook Circulación general

import numpy as np 
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
%matplotlib inline
```

```python
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Tiempo (s)', ylabel='Altura del mar (m)',
       title='Prueba de gráfico')
ax.grid()

fig.savefig("test.png")
plt.show()
```
